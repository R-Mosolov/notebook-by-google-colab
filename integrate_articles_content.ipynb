```
'''
Define separate tables paths
'''
df1 = pd.read_csv('/content/drive/MyDrive/Science/Datasets/cyberleninka-sociology-articles/cyberleninka-sociology-articles__2-0_articles-content/cyberleninka-sociology-articles__2-1_articles-content-25.csv')
df2 = pd.read_csv('/content/drive/MyDrive/Science/Datasets/cyberleninka-sociology-articles/cyberleninka-sociology-articles__2-0_articles-content/cyberleninka-sociology-articles__2-1-1_articles-content-3.csv')
df3 = pd.read_csv('/content/drive/MyDrive/Science/Datasets/cyberleninka-sociology-articles/cyberleninka-sociology-articles__2-0_articles-content/cyberleninka-sociology-articles__2-1-2_articles-content-6.csv')
df4 = pd.read_csv('/content/drive/MyDrive/Science/Datasets/cyberleninka-sociology-articles/cyberleninka-sociology-articles__2-0_articles-content/cyberleninka-sociology-articles__2-1-3-3_articles-content-3.csv')
```

```
'''
Slice rows with errors and empty values to get full work dataframes
'''
df1_ok = df1
df2_ok = df2
slice_df_3 = slice(5699)
df3_ok = df3[slice_df_3]
slice_df_4 = slice(2559)
df4_ok = df4[slice_df_4]
```

```
'''
Delete the unused column from dataframes 2-4 ("article_number")
'''
df2_ok.drop(["article_number"], axis = 1, inplace = True) 
df3_ok.drop(["article_number"], axis = 1, inplace = True)
df4_ok.drop(["article_number"], axis = 1, inplace = True)
```

```
'''
Create the general dataframe
'''
df = pd.concat([df1_ok, df2_ok, df3_ok, df4_ok])

df['article_link']
```

```
'''
Save general dataset into a new table
'''
# Save parsed data as an Excel file
result_file_name = (
  '/content/drive/MyDrive/Science/Datasets/cyberleninka-sociology-articles/cyberleninka-sociology'
  + '-articles__2-0_articles-content/integrated-dataframes/integrated-dataframe__actuality-08-01-2021.csv'
)
is_column_name = True

# Remove old file
try:
  os.remove(result_file_name)
except:
  pass

with open(result_file_name, 'a') as csvfile:
  writer = csv.writer(csvfile)
  
  if is_column_name == True:
    writer.writerow([
      'article_link',
      'article_views',
      'article_downloads',
      'journal_title',
      'journal_link',
      'key_words',
      'abstract',
      'article_text'
    ])
    is_column_name = False
  
  for index in range(0, len(df1_ok)):
    writer.writerow([
      df1_ok['article_link'][index],
      df1_ok['article_views'][index],
      df1_ok['article_downloads'][index],
      df1_ok['journal_title'][index],
      df1_ok['journal_link'][index],
      df1_ok['key_words'][index],
      df1_ok['abstract'][index],
      df1_ok['article_text'][index]
    ])

  for index in range(0, len(df2_ok)):
    writer.writerow([
      df2_ok['article_link'][index],
      df2_ok['article_views'][index],
      df2_ok['article_downloads'][index],
      df2_ok['journal_title'][index],
      df2_ok['journal_link'][index],
      df2_ok['key_words'][index],
      df2_ok['abstract'][index],
      df2_ok['article_text'][index]
    ])

  for index in range(0, len(df3_ok)):
    writer.writerow([
      df3_ok['article_link'][index],
      df3_ok['article_views'][index],
      df3_ok['article_downloads'][index],
      df3_ok['journal_title'][index],
      df3_ok['journal_link'][index],
      df3_ok['key_words'][index],
      df3_ok['abstract'][index],
      df3_ok['article_text'][index]
    ])

  for index in range(0, len(df4_ok)):
    writer.writerow([
      df4_ok['article_link'][index],
      df4_ok['article_views'][index],
      df4_ok['article_downloads'][index],
      df4_ok['journal_title'][index],
      df4_ok['journal_link'][index],
      df4_ok['key_words'][index],
      df4_ok['abstract'][index],
      df4_ok['article_text'][index]
    ])

```
